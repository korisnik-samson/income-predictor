```python
# AGE
plt.figure(figsize=(16, 8))

sns.set_theme(style='darkgrid')
sns.set_palette('magma')
sns.histplot(data=data, x='age', hue='income', bins=20, multiple='stack')

plt.xlabel('AGE')
plt.ylabel('COUNT')
plt.title('AGE DISTRIBUTION')

plt.show()
```

```python
# WORKCLASS

data.groupby('workclass').size()
```

```python
unknown_workclass = data.loc[data['workclass'] == '?']
print('**AGE DIST FOR WORKCLASS "?"\n', unknown_workclass['age'].describe())

plt.figure(figsize=(16, 8))
plt.title('AGE DISTRIBUTION FOR WORKCLASS "?"')
plt.hist(unknown_workclass['age'], bins=20)

sns.histplot(data=data.loc[data['workclass'] == '?'], x='age', hue='income', bins=20, multiple='stack')
```
